```python
import numpy as np
import cv2 as cv
#allow acces to gg drive
from google.colab.patches import cv_imshow
from google.colab import drive
```

```python
#load image
#To read the image, we use the ‚Äúimread‚Äù method from the cv2 package, where the first parameter is the image‚Äôs path, including filename and extension,
# and the second parameter is a flag that determines how to read in the image
image_path='/content/drive/MyDrive/Colab Notebooks/mrbean.jpg'
image=cv.imread(image_path,0)
```

```python
#create kernel: matrix size 5x5
kernel1= np.ones((5,5),np.float32)/25.0
kernel2= np.ones((9,9),np.float32)/81.0
```

```python
#compute mean for each pixel
image_filter5x5= cv.filter2D(image, cv.CV_8U,kernel1)
image_filter9x9= cv.filter2D(image, cv.CV_8U,kernel2)
```

```python
#display 
print("original picture:")
cv_imshow(image)
print("\nfiltered picture with kernel 5x5:")
cv_imshow(image_filter5x5)
print("\nfiltered picture with kernel 9x9:")
cv_imshow(image_filter9x9)
```

a) When we increase the kernel size from 5x5 to 9x9, we can see that the image is getting more and more blurry. Kernel is a convolution matrix used for blurring, sharpening,.. For each 5x5 block of pixels in the original image, we multiply each pixel by the corresponding entry of kernel and then we take the sum. That sum becomes new pixel on the filtered images.
